```python
import requests
from bs4 import BeautifulSoup

url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver'
res = requests.get(url)

print(res) # 응답 객체
# res.text # HTML 텍스트


# 첫번째 인자를 텍스트로 넣어줘야 함
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
titles = soup.find_all(class_ = "title")
```

```python
titles = titles[:10]
```

```python
len(titles)
```

```python
title_names = soup.select('td.title a')
```

```python
title_names = title_names[:10]
titles_info_list = list()



for title_name in title_names:
    title_link = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver' + title_name['href']
    response_title = requests.get(title_link)
    soup_title = BeautifulSoup(response_title.text, 'html.parser')

    title_info_list = list()

    # .1 제목
    title_info_list.append(title_name.text)

    # 2. 관람객 평점
    enums = soup_title.select('div.score div.star_score em')[:4]
    score = 0
    divider = 1
    for enum in enums:
        if(enum.text == '.'):
            continue
        score += int(enum.text) / divider
        divider *= 10

    title_info_list.append(format(score, '.2f'))


    # 3. 개요
    if(soup_title.select_one('dl.info_spec a')):
        title_info_list.append(soup_title.select_one('dl.info_spec a').text)


    # 4. 등급
    if(soup_title.select_one('dt.step4+dd a')):
        title_info_list.append(soup_title.select_one('dt.step4+dd a').text)


    # 5. 흥행
    if(soup_title.select_one('div.step9_cont p.count')):
        title_info_list.append(soup_title.select_one('div.step9_cont p.count').text)

    titles_info_list.append(title_info_list)

```
